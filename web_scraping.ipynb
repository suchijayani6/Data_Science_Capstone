```
df = pd.read_html(url, header =0)
test = df[0]
test.head()
```

```
from bs4 import BeautifulSoup
import requests
```

```
test2 = test[test.Borough != 'Not assigned']
test2
```

```
test3 = test2.groupby(['Postcode', 'Borough'])['Neighbourhood'].apply(', '.join).reset_index()
test3
```

```
final = (test3['Neighbourhood']=='Not assigned')
test3.loc[final, 'Neighbourhood'] = test2['Borough']
test3
```

```
import numpy as np  
import pandas as pd
```

```
geo = pd.read_csv('https://cocl.us/Geospatial_data')
geo.head ()
```

```
result = pd.merge(left=test3, right=geo, how='left', left_on='Postcode', right_on='Postal Code')

result.head ()
```

```
result = result.drop ('Postal Code', axis = 1)
result
```

```
result.head()
```

```
address = 'Toronto, Canada '

geolocator = Nominatim(user_agent="ny_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto City are {}, {}.'.format(latitude, longitude))
```

```
# create map of New York using latitude and longitude values
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, borough, neighbourhood in zip(result['Latitude'], result['Longitude'], result['Borough'], result['Neighbourhood']):
    label = '{}, {}'.format(neighbourhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```
